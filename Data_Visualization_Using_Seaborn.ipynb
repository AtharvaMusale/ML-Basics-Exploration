## FacetGrid
* MultiPlot Grid for plotting conditional relationship
* https://seaborn.pydata.org/generated/seaborn.FacetGrid.html

```python

#load data
from tabulate import tabulate
import seaborn as sns

tips = sns.load_dataset("tips")

#Ref: https://pypi.org/project/tabulate/
#print(tabulate(tips, headers='keys', tablefmt='github'))
print(tips.head(10))
```

```python
# Plot the data for 
sns.set(style="ticks", color_codes=True)
g = sns.FacetGrid(tips, col="time", row="smoker") #time and smoker are categorical
print(type(g))
```

```python
# Ref: https://seaborn.pydata.org/generated/seaborn.FacetGrid.map.html#seaborn.FacetGrid.map
g.map(plt.hist, "total_bill")
```

```python
import matplotlib.pyplot as plt
g.map(plt.hist, "total_bill")
```

```python
import numpy as np

bins = np.arange(0, 65, 3)
g = sns.FacetGrid(tips, col="time",  row="smoker")
g.map(plt.hist, "total_bill", bins=bins)


#Conclusion?
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.scatter, "total_bill", "tip", edgecolor="w" )
```

```python
g = sns.FacetGrid(tips, col="time",  hue="smoker")
#print(type(g.map(plt.scatter, "total_bill", "tip", edgecolor="w")))
g = (g.map(plt.scatter, "total_bill", "tip", edgecolor="w")).add_legend()
```

```python
data = sns.load_dataset('iris')
print(data.head())
print(data.columns)
```

# **ScatterPlot**

Observations - 
* Setosa petal_width and petal_length are significantly lower and can be distinguished easily from other two classes. So these features can be given high weightage while seperating the Setosa flowers from the rest of the types.

```python
sns.set(style='whitegrid')
sns.scatterplot(x = 'sepal_width', y = 'species', data=data, hue='species',style='species',palette='deep',legend='full',hue_norm=(0,10))
```

```python
sns.set(style='whitegrid')
sns.scatterplot(x = 'petal_width', y = 'species', data=data, hue='species',style='species',palette='deep',legend='full',hue_norm=(0,10))
```

# **PairPlot**

```python
plt.close();
sns.set_style("whitegrid");
sns.pairplot(data, hue="species", size=3);
plt.show()
```

**Observations**
1. petal_length and petal_width are the most useful features to identify various flower types.
2. While Setosa can be easily identified (linearly seperable), Virginica and Versicolor have some overlap (almost linearly seperable).
3. We can find "lines" and "if-else" conditions to build a simple model to classify the flower types.

```python
# What about 1-D scatter plot using just one feature?
#1-D scatter plot of petal-length
import numpy as np
data_setosa = data.loc[data["species"] == "setosa"];
data_virginica = data.loc[data["species"] == "virginica"];
data_versicolor = data.loc[data["species"] == "versicolor"];
#print(data_setosa["petal_length"])
plt.plot(data_setosa["petal_length"], np.zeros_like(data_setosa['petal_length']), 'o')
plt.plot(data_versicolor["petal_length"], np.zeros_like(data_versicolor['petal_length']), 'o')
plt.plot(data_virginica["petal_length"], np.zeros_like(data_virginica['petal_length']), 'o')

plt.show()
#Disadvantages of 1-D scatter plot: Very hard to make sense as points 
#are overlapping a lot.
#Are there better ways of visualizing 1-D scatter plots?
```
